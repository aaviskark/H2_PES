```python
# X and Y variables
x = np.linspace(0.3,2.0,18)
y = [1267.197336, 521.2389469, 192.5911489, 50.77158487, 1.372032565, 0, 23.94125675, 60.8847918, 103.8928851, 148.9641207, 193.7700761, 236.9425256, 277.7025628, 315.6371707,350.5872966, 382.5462194, 411.6014204, 437.9007154]
# Variables for Experimental Val.
a = [0.741]
b = [0]
```
